```python
flt_files = glob.glob('*_flt.fits')
```

```python
file = fits.open(flt_files[0])
file[0].header
```

```python
for img_files in flt_files:
    updatewcs.updatewcs(img_files, use_db=False)
```

```python
lst_f150lp = []
lst_f165lp = []
for flt_file in flt_files:
    file = fits.open(flt_file)
    filt = file[0].header['FILTER1']
    if filt == 'F150LP':
        lst_f150lp.append(flt_file)
    elif filt == 'F165LP':
        lst_f165lp.append(flt_file)
```

```python
astrodrizzle.AstroDrizzle(lst_f150lp, 
                              output = 'F150LP_total', 
                              clean = True,
                              build = True,
                              context = True,
                              runfile = 'F150LP_total',
                              num_cores = 1,
                              in_memory = False,
                              resetbits = 4096,
                              median = False,
                              static = False,
                              skysub = False,
                              driz_separate = False,
                              blot = False,
                              driz_cr = False,
                              final_bits = 0,
                              final_pixfrac = 0.6,
                              final_outnx=1947,
                              final_outny=1960, 
                              final_rot = 0.0,
                              final_scale = 0.025,
                              final_wcs = True)
```
